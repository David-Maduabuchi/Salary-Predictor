```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("salaries.csv")
df.head()
```

```python
x = df.iloc[:, :-1].values
x[0:5]
```

```python
y = df.iloc[:, -1].values
y[0:5]
```

```python
plt.scatter(x, y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)

```

from sklearn.linear_model import LinearRegression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.predict([[7], [5], [3.5], [8]])
# This is the predicted salary for seven years of experience, 5 years and 3.5 years
```

```python
# So we have to show the values for the predictions we are going to do, for the
y_pred = model.predict(x_test)
y_pred
# x_test
```

```python
error = y_pred - y_test
error
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)

# Si= so we test for the accuracy by pitting the test against the predictions
print(f"The accuracy of the model is approximately: {r2 *100:.2f}%")
```
